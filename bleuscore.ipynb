```python
import torch
from transformers import MarianMTModel, MarianTokenizer
```

```python
model_name = "Helsinki-NLP/opus-mt-hi-en"
model = MarianMTModel.from_pretrained(model_name)
tokenizer = MarianTokenizer.from_pretrained(model_name)
```

```python
# Prepare data and tokenize
source_text = "यह एक उदाहरण है।"
inputs = tokenizer.encode(">>hi<<" + source_text, return_tensors="pt", max_length=512, truncation=True)
```

```python
translation_ids = model.generate(inputs, max_length=150, num_beams=4, length_penalty=2.0, early_stopping=True)
translated_text = tokenizer.decode(translation_ids[0], skip_special_tokens=True)
```

```python
print(translated_text)
```

```python
from nltk.translate.bleu_score import corpus_bleu
```

```python
# Example reference and candidate translations
references = [["This is an example."]]
candidates = ["This is an example as well."]
```

```python
# Calculate BLEU score
bleu_score = corpus_bleu(references, candidates)
print("BLEU Score:", bleu_score)
```
